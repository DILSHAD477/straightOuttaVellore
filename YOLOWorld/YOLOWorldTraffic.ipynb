# Video Inference

```python
model.set_classes(["car","bike","truck","autorickshaw","ambulance","bus","person"])
```

```python
VIDEO_PATH = r"C:\Users\get2b\Desktop\Arav\Hackathons\Karnataka State Police\New Dataset\Video to test on\Video12.mp4"
video_info = sv.VideoInfo.from_video_path(VIDEO_PATH)
```

```python
def process_frame(frame: np.ndarray, _) -> np.ndarray:
    results = model(frame, imgsz=480)[0]
    
    detections = sv.Detections.from_ultralytics(results)

    box_annotator = sv.BoxAnnotator(thickness=4, text_thickness=4, text_scale=1)

    labels = [f"{model.names[class_id]} {confidence:0.2f}" for _, _, confidence, class_id, _, _ in detections]
    frame = box_annotator.annotate(scene=frame, detections=detections, labels=labels)

    return frame

sv.process_video(source_path=VIDEO_PATH, target_path=r"C:\Users\get2b\Desktop\Arav\Hackathons\Karnataka State Police\result.mp4", callback=process_frame)
```
